```python
#function
def fun():
    print("welcome")
fun()
```

```python
#argument passing
def evenodd (x):
    if x% 2==0:
        print("even")
    else:
        print("odd")
evenodd(10)
evenodd(15)
```

```python
#default argument
def myfun(x,y=20):
    print("x:",x)
    print("y:",y)
myfun(30)
```

```python
#key argument
def student (fname, lname):
    print(fname, lname)

student(fname="Practice", lname="Daily")
student(fname="Daily", lname="Practice")
```

```python
#positional argument
def person(name,age):
    print("Hi I am",name)
    print("My age is",name)

#case1
person("Ram", 25)

#case2
person("Ram", 25)
```

```python
#function within function
def f1():
    s="I love you"
    def f2():
        print(s)
    f2()
f1()    
```

```python
#pass by refernce
def myfun(x):
    x[0]=20
lst=[10,20,30,40,50]
print(lst)
myfun(lst)
print(lst)
```

```python
#pass by value
def myFun(x):
    x=20
a=20
print(a)
myFun(a)
print(a)
#a is not modified
```

```python
#file handling
f = open("kk greek.txt","w")
f.write("File Handling Examples")
f.close()
```

```python
#opening file
f=open("kk greek.txt", "r")
print(f)
```

```python
## closing file
file=open("kk greek.txt","r")
#perform file opertions
file.close()
```

```python
#checking file properties
f = open("kk greek.txt", "r")
print("Filename:", f.name)
print("Mode:", f.mode)
print("Is closed?:", f.closed)

f.close()
print("Is closed?", f.closed)
```

```python
#reading file
f = open("kk greek.txt", "r")
content= f.read()
print(content)
f.close()
```

```python
#writing file
with open("kk greek.txt", "w") as file:
    file.write("Hello Python!\n")
    file.write("File Handling is easy with Python.")
print("File written successfully")
```
